```python
# 1. What is the average median income of the data set and check the distribution of data using appropriate plots.
# Please explain the distribution of the plot.

average_income = data['median_income'].mean()  # .mean() is used to calculate average value of the median_income column

print("Average of median_income :",average_income)

data.hist(bins=30,figsize=(20,15),color='orange') # Histogram helps us to see the distribution of a numerical values
plt.show()
```

```python
# 6. Write a programming construct (create a user defined function) to 
# calculate the median value of the data set wherever required.

import re
import statistics

# Sample dataset
dataset = {
    "column_1": data['longitude'],"column_2": data['latitude'],"column_3": data['housing_median_age'],
    "column_4": data['total_rooms'],"column_5": data['total_bedrooms'],"column_6": data['population'],
    "column_7": data['households'],"column_8": data['median_income'],"column_9": data['median_house_value']
}

def calculate_median(data):  # defining a function calculate_median
    return statistics.median(data)

# Iterate through columns in the dataset
for column_name, column_data in dataset.items():
    # Check if the column name contains a digit using regex
    if re.search(r'\d', column_name):  # checks if there is atleast one digit in the column
        median = calculate_median(column_data)
        print((column_name,median))
```

```python
# 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. 
# If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

initial_count = 0  # initializing count value to zero

count_of_bedrooms = [
              (data['total_bedrooms'] <=10),
              (data['total_bedrooms'] >=11) & (data['total_bedrooms'] <=1000),
              (data['total_bedrooms'] > 1000)
             ]    

# list of conditions are provided for the given dataset

sizes_of_bedrooms = ['small','medium','large']  # categorizing the sizes

data['total_bedroom_size'] = np.select(count_of_bedrooms,sizes_of_bedrooms)  # new column is created

# selects the category size for given count and assigns that value for each row in a column

data
```
